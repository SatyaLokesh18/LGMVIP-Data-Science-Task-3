```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Print feature and target names
print("Features:", iris.feature_names)
print("Targets:", iris.target_names)

# Remove specific samples from the dataset
removed_samples = [0, 50, 100]
new_target = np.delete(iris.target, removed_samples)
new_data = np.delete(iris.data, removed_samples, axis=0)

# Train the Decision Tree Classifier
classifier = DecisionTreeClassifier()
classifier.fit(new_data, new_target)

# Predict the labels for the removed samples
predicted_labels = classifier.predict(iris.data[removed_samples])

# Print original and predicted labels
print("Original Labels:", iris.target[removed_samples])
print("Predicted Labels:", predicted_labels)

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(classifier, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```
